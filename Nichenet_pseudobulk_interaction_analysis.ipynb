```r
# read in exp data and metadata
tnk_gene_exp <- read.csv("pseudobulk_Cd4notcd8_expressing_t_cells.csv",row.names=1)
metadata <- read.csv("pseudobulk_Cd4notcd8_expressing_t_cells_meta.csv", row.names=1)

# take just cd8a and Igg and just cd4 expressing
metadata <- metadata[metadata$gran_cluster %in% c('Cluster 2','Cluster 4','Cluster 3'),]
metadata <- metadata[metadata$treatment %in% c("IgG", "CD8a"),]
metadata <- metadata[metadata$CD4.not.CD8=="True",]
# filter out genes not expressed highly enough based on the thereshold defined above
tnk_gene_exp<-tnk_gene_exp[rownames(metadata),colnames(tnk_gene_exp) %in% expressed_genes_receiver] 
expressed_genes_receiver <- colnames(tnk_gene_exp)

tnk_gene_exp <- na.omit(tnk_gene_exp)
tnk_gene_exp <- t(tnk_gene_exp)
metadata$M.Number <- factor(metadata$M.Number)
metadata$monkey_condition_categories <- factor(metadata$monkey_condition_categories)

# run deseq
dds <- DESeqDataSetFromMatrix(countData = tnk_gene_exp, colData = metadata, design = ~ factor(monkey_condition_categories)+factor(treatment))
dds_thistype_wald <- DESeq(dds,test="Wald")
res <- results(dds_thistype_wald)
write.csv(res, "differential_genes_cd4notcd8_igg_cd8a_gran_clusters.csv",quote=FALSE)

```

```r
# filter top differential genes

geneset_oi <- rownames(subset(ressub,ressub$log2FoldChange < -.5))
```

```r
# read in ligand network 
lr_network = readRDS(url("https://zenodo.org/record/3260758/files/lr_network.rds"))



ligands = lr_network %>% pull(from) %>% unique()
expressed_ligands = intersect(ligands,expressed_genes_sender)

receptors = lr_network %>% pull(to) %>% unique()
expressed_receptors = intersect(receptors,expressed_genes_receiver)

# our ligand receptor network of interest
lr_network_expressed = lr_network %>% filter(from %in% expressed_ligands & to %in% expressed_receptors) 
potential_ligands = lr_network_expressed %>% pull(from) %>% unique()
```

```r
background_expressed_genes = expressed_genes_receiver %>% .[. %in% rownames(ligand_target_matrix)]
head(background_expressed_genes)
```

```r
ligand_activities = predict_ligand_activities(geneset = geneset_oi, background_expressed_genes = background_expressed_genes, ligand_target_matrix = ligand_target_matrix, potential_ligands = potential_ligands)
```

```r
best_upstream_ligands = ligand_activities %>% top_n(30, pearson) %>% arrange(-pearson) %>% pull(test_ligand)
head(best_upstream_ligands)
```

```r
best_upstream_ligands
```

```r
active_ligand_target_links_df = best_upstream_ligands %>% lapply(get_weighted_ligand_target_links,geneset = geneset_oi, ligand_target_matrix = ligand_target_matrix, n = 250) %>% bind_rows()
# this is taking the top ligands, looking at their top 250 targets, taking the targets of those 250 that are in geneset_oi then saving the ligand, target, and the weight from the ligand_target_matrix into the DF
```

```r
write.csv(active_ligand_target_links_df,"active_ligand_target_links_cluster_subbed_CD8avIgG_cd4.csv",quote=FALSE)
```
